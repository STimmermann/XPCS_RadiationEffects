```python
#measurement list is csv file with columns:
#'measurement': names of measurements, usually m0XY...
# ser_start: first series to analyse
# ser_end: first series that is not analzed anymore
dataSets = pd.read_csv('measurement_list.csv', #path to csv file
                       delimiter=';', # delimiter in csv file
                       header=0, # first row is header
                      )
```

## SAXS analysis

```python
setupfile = 'setupfiles/USAXS_e4m_12q.pkl' # name of setup file

setup = open(setupfile,'rb')
setup_dict = pkl.load(setup)
setup.close()

for m in [10]:  # range/list of datasets that are analysed, number = index in print(datasets)
    meas = dataSets['measurement'][m] # name of dataset
    for ser in range(int(dataSets['ser_start'][m]),int(dataSets['ser_end'][m])): # loop over series
        # connect Xana to all series,
        # add ../ before raw depending on where the raw data is, change e4m to e500 for small EIGER
        d = Xana(sample=meas, fmtstr='p10_eiger_h5',setupfile=setupfile,) # create Xana object
        d.mksavdir(str(meas), './')      #create directory for saving, here name = name of measurement + extension
        d.connect('../../raw/'+meas+'_{:05d}/'.format(ser)+'/e4m') # connect Xana object to series         
        firstframe = 0 # first frame to be analysed
        stepframe = 100 # how many frames are averaged in one SAXS curve
        nframes = d.meta['nframes'][0]-stepframe # number of frames
        for first_ind in np.arange(firstframe,nframes,stepframe): # loop over chunks of frames
            d.analyze(0, # give the index of the dataset that Xana is connected to, as we connect only one, this is always 0
                      # the multiprocessing is sometimes getting stuck so connecing only one series at a time is safer
                      'saxs', # analysis mode 'saxs' instead of 'xpcs'
                      verbose=False, #less detailed output
                      first = int(first_ind), #first frame of chunck
                      last=int(first_ind+stepframe), # last frame of chunck
                     ) 
```

## XPCS analysis

```python
setupfile = 'setupfiles/USAXS_e4m_12q.pkl' # name of setup file

setup = open(setupfile,'rb')
setup_dict = pkl.load(setup)
setup.close()


for m in [10]: # range/list of datasets that are analysed, number = index in print(datasets)
    meas = dataSets['measurement'][m] # name of dataset
    
    for ser in range(int(dataSets['ser_start'][m]),int(dataSets['ser_end'][m])): # loop over series
        d = Xana(sample=meas, fmtstr='p10_eiger_h5',setupfile=setupfile,) # create Xana object
        d.mksavdir(str(meas), './')      #create directory for saving, here name = name of measurement
        # connect Xana to single series,
        # add ../ before raw depending on where the raw data is, change e4m to e500 for small EIGER
        d.connect('../../raw/'+meas+'_{:05d}/'.format(ser)+'/e4m') # connect Xana object to series         
        
        d.analyze(0, # give the index of the dataset that Xana is connected to, as we connect only one, this is always 0
                      # the multiprocessing is sometimes getting stuck so connecing only one series at a time is safer
                  'xpcs', # analysis mode 'xpcs' instead of 'saxs'
                  saxs=None, # saxs=true does some normalization according to static scattering
                  verbose=False, # less output
                  twotime_par=range(np.multiply(len(setup_dict.qv_init),len(setup_dict.phiv_init))), # calculate TTC for all q + phi pairings defined in setupfile
                  nprocs=5, # number of processes for multiprocessing 
                  chunk_size=20, # size of chunks for multiprocessing
                  nread_procs=1,
                  #tt_max_images=3000, #maximum images for TTC
                  norm='symmetric', # normalization for correlation functions
                  #first=10, #first frame that is included (sometimes the first frames are without beam)
                  #last = , #last frame to be considered 
                  ) 
```

# Detailed intensity analysis

```python
t_exp_till_masterdose = {'0':75,'1':103,'2':135,'4':265,'6':491,'8':909,'12':3333} #time in seconds for each absorber until a certain dose is reached
```

```python
logbook = pd.read_csv('egg_logbook.csv', delimiter=';',header=0,) # logbook file mit meas_id, measurement, temp(erature), ser(ies), ab(sorber)
```

```python
measurement_no = 10 # choose measurement from measurement list
meas = dataSets['measurement'][measurement_no] # name of dataset
sublog = logbook.loc[logbook.meas_id == measurement_no] # look up series and absorbers in logbook file
print(sublog)
```

```python
setupfile = 'setupfiles/USAXS_e4m_12q.pkl' # name of setup file

setup = open(setupfile,'rb')
setup_dict = pkl.load(setup)
setup.close()

series_list =[4,5,7,9,11,13,15]#list of series for that you want a more detailed intensity analysis
for ser in series_list:
    d = Xana(sample=meas, fmtstr='p10_eiger_h5',setupfile=setupfile,) # create Xana object
    d.mksavdir(str(meas)+'_Intensity_300kGy', './')      #create directory for saving, here name = name of measurement + extension
    # connect Xana to series,
    # add ../ before raw depending on where the raw data is, change e4m to e500 for small EIGER
    d.connect('../../raw/'+meas+'_{:05d}/'.format(ser)+'e4m') # connect Xana object to series
    t_exp = d.meta['t_exposure'][0] # get exposure time from Xana data base
    nframes = d.meta['nframes'][0] # get number of frames from Xana data base
    ab_df = sublog.loc[sublog['ser'] == ser] #get absorber value for series
    ab = ab_df['ab'].values[0]
    t_exp_max = t_exp_till_masterdose[str(ab)] # get exposure time from dictionary defined above
    max_frames = np.floor(t_exp_max/t_exp) # transform into number of frames with the exposure time
    
    #if number of frames is smaller than max_frames, use all frames instead
    use_frames = 0
    if max_frames > nframes:
        use_frames = nframes
    else:
        use_frames = max_frames
        
    step = int(np.ceil(use_frames/100)) # calculate 100 chunks of intensity
    for first_ind in np.arange(0,use_frames-step,step): # loop over chunks
        d.analyze(0, # give the index of the dataset that Xana is connected to, as we connect only one, this is always 0
                  # the multiprocessing is sometimes getting stuck so connecing only one series at a time is safer
                  'saxs', # analysis mode 'saxs' instead of 'xpcs'
                  verbose=False, #less detailed output
                  first = int(first_ind), #first frame of chunck
                  last=int(first_ind+stepframe), # last frame of chunck
                 ) 
```
